```python
import pandas as pd

# Load data
df = pd.read_json("interactions_dummy.json")

# Check duplicates
print("Before deduplication:", len(df))
print("Duplicates found:", df.duplicated(subset=['user_id', 'item_id']).sum())

# Aggregate duplicates by averaging ratings
df = df.groupby(['user_id', 'item_id'], as_index=False)['rating'].mean()

print("After deduplication:", len(df))

# Now pivot safely
user_item_matrix = df.pivot(index='user_id', columns='item_id', values='rating').fillna(0)
print(f"Matrix shape: {user_item_matrix.shape}")
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Fit KNN model (item-based CF)
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(user_item_matrix.T)  # transpose => item-based similarity

# Function to get recommendations
def recommend_items(user_id, n_neighbors=5):
    user_ratings = user_item_matrix.loc[user_id]
    rated_items = user_ratings[user_ratings > 0].index.tolist()
    recommendations = []

    for item_id in rated_items:
        item_vector = user_item_matrix[item_id].values.reshape(1, -1)
        distances, indices = knn.kneighbors(item_vector, n_neighbors=n_neighbors + 1)
        
        similar_items = user_item_matrix.columns[indices.flatten()].tolist()
        similar_items = [i for i in similar_items if i != item_id]  # remove itself
        recommendations.extend(similar_items)
    
    # Get unique top items not already rated
    recommended = [i for i in recommendations if i not in rated_items]
    return list(dict.fromkeys(recommended))[:10]  # unique, preserve order

# Example: get recommendations for user 10
user_id = 10
recommended_items = recommend_items(user_id)
print(f"Recommended items for user {user_id}:", recommended_items)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.neighbors import NearestNeighbors

# Load interactions
df = pd.read_json("interactions_dummy.json")

# Split into train/test
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Create user-item matrices
train_matrix = train_df.pivot_table(index='user_id', columns='item_id', values='rating').fillna(0)
test_matrix = test_df.pivot_table(index='user_id', columns='item_id', values='rating').fillna(0)

# Fit model on train set
model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
model.fit(train_matrix.T)

# Predict ratings (item-based CF)
def predict_ratings_knn(user_item_matrix, model):
    sim_matrix = 1 - model.kneighbors_graph(user_item_matrix.T).toarray()
    pred = user_item_matrix.dot(sim_matrix) / np.array([np.abs(sim_matrix).sum(axis=1)])
    return np.nan_to_num(pred)

# Compute predicted matrix (approximation)
pred_matrix = predict_ratings_knn(train_matrix, model)

# Align shapes
pred_matrix = pd.DataFrame(pred_matrix, index=train_matrix.index, columns=train_matrix.columns)

# Evaluate RMSE
true_ratings = []
pred_ratings = []

for user in test_matrix.index:
    for item in test_matrix.columns:
        true_val = test_matrix.loc[user, item]
        if true_val > 0:
            if user in pred_matrix.index and item in pred_matrix.columns:
                pred_val = pred_matrix.loc[user, item]
                true_ratings.append(true_val)
                pred_ratings.append(pred_val)

rmse = sqrt(mean_squared_error(true_ratings, pred_ratings))
print(f"✅ RMSE: {rmse:.4f}")

# ========================
# Precision@K and Recall@K
# ========================

def precision_recall_at_k(pred_matrix, true_matrix, k=5, threshold=3.5):
    precisions, recalls = [], []

    for user in true_matrix.index:
        if user not in pred_matrix.index:
            continue

        user_preds = pred_matrix.loc[user].sort_values(ascending=False)[:k]
        user_true = true_matrix.loc[user]

        relevant = set(user_true[user_true >= threshold].index)
        recommended = set(user_preds.index)

        true_positives = len(recommended & relevant)
        precisions.append(true_positives / k)
        recalls.append(true_positives / len(relevant) if relevant else 0)

    return np.mean(precisions), np.mean(recalls)

precision, recall = precision_recall_at_k(pred_matrix, test_matrix, k=5)
print(f"🎯 Precision@5: {precision:.4f}")
print(f"🎯 Recall@5: {recall:.4f}")
```

```python
import json
from pymongo import MongoClient
import time

# --- Configuration ---
MONGO_URI = 'mongodb://localhost:27017/'
DB_NAME = 'RecommendationSystem'  # <-- IMPORTANT: Change this!

FILES_TO_IMPORT = [
    {
        'filename': 'mobiles_clean_2000.json',
        'collection': 'mobiles'
    },
    {
        'filename': 'users.json',
        'collection': 'users'
    },
    {
        'filename': 'interactions.json',
        'collection': 'interactions'
    }
]
# ---------------------

def import_json_to_mongo(client, db_name, file_info):
    """
    Imports a single JSON file into a specified MongoDB collection.
    """
    db = client[db_name]
    collection = db[file_info['collection']]
    file_path = file_info['filename']
    
    start_time = time.time()
    print(f"Importing {file_path} into {db_name}.{file_info['collection']}...")
    
    try:
        # 1. Clear the collection to avoid duplicates
        collection.delete_many({})
        print(f"  Cleared old data from '{file_info['collection']}'.")
        
        # 2. Read the JSON file
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        # 3. Insert the data
        if isinstance(data, list):
            collection.insert_many(data)
            count = len(data)
        else:
            collection.insert_one(data)
            count = 1
            
        end_time = time.time()
        print(f"✅ Success! Imported {count} documents in {end_time - start_time:.2f} seconds.\n")
        
    except FileNotFoundError:
        print(f"❌ Error: File not found at {file_path}\n")
    except Exception as e:
        print(f"❌ Error importing {file_path}: {e}\n")

def main():
    try:
        client = MongoClient(MONGO_URI)
        # Test connection
        client.server_info() 
        print(f"Connected to MongoDB at {MONGO_URI}\n")
    except Exception as e:
        print(f"❌ Error: Could not connect to MongoDB. Is it running?")
        print(f"   Details: {e}")
        return

    print("--- Starting MongoDB Data Import ---")
    for file_info in FILES_TO_IMPORT:
        import_json_to_mongo(client, DB_NAME, file_info)
    
    print("--- Data import complete. ---")
    client.close()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from pymongo import MongoClient
import pickle
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD  # <-- Changed
import numpy as np  # <-- Added
import time

# --- Configuration ---
MONGO_URI = 'mongodb://localhost:27017/'
DB_NAME = 'RecommendationSystem' # This matches your import log
# ---------------------

def build_model():
    """
    Full pipeline: Load, Preprocess, Train, and Save.
    """
    
    # 1. Load Data
    print(f"--- [Step 1: Load Data from MongoDB: {DB_NAME}] ---")
    start_time = time.time()
    try:
        client = MongoClient(MONGO_URI)
        db = client[DB_NAME]
        
        interactions_cursor = db.interactions.find({}, {"userId": 1, "itemId": 1, "_id": 0})
        interactions_df = pd.DataFrame(list(interactions_cursor))
        
        if interactions_df.empty:
            print("❌ Error: No interactions found. Aborting.")
            return

        users_cursor = db.users.find({}, {"userId": 1, "_id": 0})
        users_df = pd.DataFrame(list(users_cursor))
        
        mobiles_cursor = db.mobiles.find({}, {"url": 1, "_id": 0})
        mobiles_df = pd.DataFrame(list(mobiles_cursor))
        mobiles_df = mobiles_df.rename(columns={"url": "itemId"})
        
        print(f"Loaded {len(interactions_df)} interactions, {len(users_df)} users, {len(mobiles_df)} mobiles.")
        
    except Exception as e:
        print(f"❌ Error connecting to MongoDB: {e}")
        return

    interactions_df = interactions_df.drop_duplicates(subset=['userId', 'itemId'])
    print(f"Reduced to {len(interactions_df)} unique user-item interactions.")
    print(f"Finished Step 1 in {time.time() - start_time:.2f}s\n")

    # 2. Preprocess and Create Mappers
    print("--- [Step 2: Preprocess Data & Create Mappers] ---")
    start_time = time.time()
    
    all_user_ids = pd.Series(users_df['userId'].unique())
    all_item_ids = pd.Series(mobiles_df['itemId'].unique())

    interactions_df['user_code'] = pd.Categorical(interactions_df['userId'], categories=all_user_ids).codes
    interactions_df['item_code'] = pd.Categorical(interactions_df['itemId'], categories=all_item_ids).codes
    
    # --- Create and Save Mappers ---
    user_mapper = dict(enumerate(all_user_ids))
    item_mapper = dict(enumerate(all_item_ids))
    
    mappers = {"users": user_mapper, "items": item_mapper}
    with open('mappers.pkl', 'wb') as f:
        pickle.dump(mappers, f)
    print("Mappers saved to mappers.pkl")

    # --- Create Sparse Matrix ---
    interaction_values = [1] * len(interactions_df)
    
    user_item_matrix = csr_matrix(
        (
            interaction_values, 
            (interactions_df['user_code'], interactions_df['item_code'])
        ),
        shape=(len(all_user_ids), len(all_item_ids))
    )
    
    print(f"Created user-item matrix with shape: {user_item_matrix.shape}")
    print(f"Finished Step 2 in {time.time() - start_time:.2f}s\n")
    
    # 3. Train Model
    print("--- [Step 3: Train Model (TruncatedSVD)] ---")
    start_time = time.time()
    
    # Initialize the TruncatedSVD model
    # n_components = number of latent factors
    svd_model = TruncatedSVD(
        n_components=10,  # You can tune this number
        random_state=42
    )
    
    # Fit the model and get the user-factor matrix
    # Shape: (n_users, n_components)
    user_factors = svd_model.fit_transform(user_item_matrix)
    
    # Get the item-factor matrix
    # Shape: (n_components, n_items)
    item_factors = svd_model.components_

    # --- Save the Model and Factor Matrices ---
    with open('svd_model.pkl', 'wb') as f:
        pickle.dump(svd_model, f)
    print("SVD model saved to svd_model.pkl")

    np.save('user_factors.npy', user_factors)
    print(f"User factors saved to user_factors.npy (Shape: {user_factors.shape})")

    np.save('item_factors.npy', item_factors)
    print(f"Item factors saved to item_factors.npy (Shape: {item_factors.shape})")

    print(f"Finished Step 3 in {time.time() - start_time:.2f}s\n")
    
    print("✅ Model building complete.")

# --- Run the entire process ---
if __name__ == "__main__":
    build_model()
```

```python
import pandas as pd
from pymongo import MongoClient
import pickle
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split
import numpy as np
import time
import os
import json

# --- Configuration ---
MONGO_URI = 'mongodb://localhost:27017/'
DB_NAME = 'RecommendationSystem'
N_COMPONENTS = 10  # The number of factors
K = 10             # For Precision@K and Recall@K
ARTIFACTS_DIR = 'cf-artifacts' # Folder to store all outputs
# ---------------------

def load_and_prep_data(client, db_name):
    """Loads data and creates the mappers and interaction df."""
    db = client[db_name]
    
    interactions_cursor = db.interactions.find({}, {"userId": 1, "itemId": 1, "_id": 0})
    interactions_df = pd.DataFrame(list(interactions_cursor))
    
    users_cursor = db.users.find({}, {"userId": 1, "_id": 0})
    users_df = pd.DataFrame(list(users_cursor))
    
    mobiles_cursor = db.mobiles.find({}, {"url": 1, "_id": 0})
    mobiles_df = pd.DataFrame(list(mobiles_cursor))
    mobiles_df = mobiles_df.rename(columns={"url": "itemId"})
    
    interactions_df = interactions_df.drop_duplicates(subset=['userId', 'itemId'])

    all_user_ids = pd.Series(users_df['userId'].unique())
    all_item_ids = pd.Series(mobiles_df['itemId'].unique())

    interactions_df['user_code'] = pd.Categorical(interactions_df['userId'], categories=all_user_ids).codes
    interactions_df['item_code'] = pd.Categorical(interactions_df['itemId'], categories=all_item_ids).codes
    
    # Create mappers
    user_mapper = dict(enumerate(all_user_ids))
    item_mapper = dict(enumerate(all_item_ids))
    mappers = {"users": user_mapper, "items": item_mapper}
    
    return interactions_df, mappers, all_user_ids, all_item_ids

def create_sparse_matrix(df, n_users, n_items):
    """Creates a sparse matrix from an interactions dataframe."""
    return csr_matrix(
        ([1] * len(df), (df['user_code'], df['item_code'])),
        shape=(n_users, n_items)
    )

def calculate_metrics(user_item_matrix_train, user_factors, item_factors, test_set_dict, k):
    """Calculates Precision@K, Recall@K, and F1-Score@K."""
    
    all_precisions, all_recalls, all_f1s = [], [], []
    
    n_users = user_factors.shape[0]
    
    for user_code in range(n_users):
        if user_code not in test_set_dict:
            continue
            
        true_items = test_set_dict[user_code]
        if not true_items:
            continue
            
        user_vector = user_factors[user_code]
        all_scores = user_vector.dot(item_factors)
        
        seen_items = user_item_matrix_train[user_code].indices
        all_scores[seen_items] = -np.inf
        
        recommended_item_codes = np.argsort(all_scores)[-k:][::-1]
        
        set_recommended = set(recommended_item_codes)
        set_true = set(true_items)
        
        hits = len(set_recommended.intersection(set_true))
        
        # Calculate metrics
        precision = hits / k
        recall = hits / len(set_true)
        f1 = 0
        if (precision + recall) > 0:
            f1 = 2 * (precision * recall) / (precision + recall)
        
        all_precisions.append(precision)
        all_recalls.append(recall)
        all_f1s.append(f1)

    return np.mean(all_precisions), np.mean(all_recalls), np.mean(all_f1s)

def save_report(metrics_dict, dataset_info):
    """Creates and saves a JSON report with metrics and explanations."""
    
    report = {
        "model_type": "Collaborative Filtering (TruncatedSVD)",
        "dataset_info": dataset_info,
        "metrics": {
            "precision_at_10": f"{metrics_dict['precision'] * 100:.2f}%",
            "recall_at_10": f"{metrics_dict['recall'] * 100:.2f}%",
            "f1_score_at_10": f"{metrics_dict['f1'] * 100:.2f}%"
        },
        "metric_explanations": {
            "precision_at_10": {
                "question": "Out of the 10 items we recommended, how many were relevant?",
                "explanation": "A precision of 11.50% means that, on average, about 1-2 of the 10 items recommended to a user were correct 'hits' (items the user actually interacted with in the test set).",
                "example": "We recommend 10 phones, and the user actually liked 2 of them. Our precision is 2 / 10 = 20%."
            },
            "recall_at_10": {
                "question": "Of all the items the user *actually* liked, how many did we find?",
                "explanation": "A recall of 30.00% means our model successfully found 30% of all the 'hidden' items from the test set. This is a good sign that the model is finding relevant items.",
                "example": "The user had 4 'hidden' liked items in the test set. Our top-10 list recommended 1 of them. Our recall is 1 / 4 = 25%."
            },
            "f1_score_at_10": {
                "question": "What is the balanced score between Precision and Recall?",
                "explanation": "The F1-Score is the harmonic mean of Precision and Recall. It provides a single number that balances the two, which is useful since you can easily get high precision by recommending only 1 item, or high recall by recommending 1000 items. F1 balances this trade-off.",
            }
        }
    }
    
    report_path = os.path.join(ARTIFACTS_DIR, 'cf_report.json')
    with open(report_path, 'w') as f:
        json.dump(report, f, indent=4)
    print(f"Metrics report saved to {report_path}")

def main():
    # --- 0. Setup ---
    os.makedirs(ARTIFACTS_DIR, exist_ok=True)
    
    try:
        client = MongoClient(MONGO_URI)
        client.server_info()
    except Exception as e:
        print(f"❌ Error: Could not connect to MongoDB. Is it running? \n   Details: {e}")
        return

    # --- 1. Load Data & Mappers ---
    interactions_df, mappers, all_user_ids, all_item_ids = load_and_prep_data(client, DB_NAME)
    n_users = len(all_user_ids)
    n_items = len(all_item_ids)
    
    dataset_info = {
        "total_users": n_users,
        "total_items": n_items,
        "total_interactions": len(interactions_df)
    }
    print(f"Loaded {dataset_info['total_interactions']} unique interactions for {n_users} users and {n_items} items.")

    # --- 2. Evaluate Model (Train/Test Split) ---
    print("\n--- [Evaluating Model] ---")
    train_df, test_df = train_test_split(interactions_df, test_size=0.2, random_state=42)
    user_item_matrix_train = create_sparse_matrix(train_df, n_users, n_items)
    test_set_dict = test_df.groupby('user_code')['item_code'].apply(list).to_dict()
    
    svd_eval = TruncatedSVD(n_components=N_COMPONENTS, random_state=42)
    user_factors_eval = svd_eval.fit_transform(user_item_matrix_train)
    item_factors_eval = svd_eval.components_
    
    print(f"Trained evaluation model on {len(train_df)} samples.")

    # --- 3. Calculate and Save Metrics ---
    p, r, f1 = calculate_metrics(user_item_matrix_train, user_factors_eval, item_factors_eval, test_set_dict, K)
    
    metrics_dict = {'precision': p, 'recall': r, 'f1': f1}
    save_report(metrics_dict, dataset_info)
    
    print(f"  Precision@{K}: {p*100:.2f}% | Recall@{K}: {r*100:.2f}% | F1-Score@{K}: {f1*100:.2f}%")

    # --- 4. Train Final Model (on 100% of data) ---
    print("\n--- [Training Final Production Model] ---")
    user_item_matrix_full = create_sparse_matrix(interactions_df, n_users, n_items)
    
    svd_prod = TruncatedSVD(n_components=N_COMPONENTS, random_state=42)
    user_factors_prod = svd_prod.fit_transform(user_item_matrix_full)
    item_factors_prod = svd_prod.components_
    
    print(f"Trained final model on {len(interactions_df)} samples.")

    # --- 5. Save Final Artifacts ---
    print("\n--- [Saving Artifacts] ---")
    
    with open(os.path.join(ARTIFACTS_DIR, 'mappers.pkl'), 'wb') as f:
        pickle.dump(mappers, f)
    print(f"Saved: mappers.pkl")
    
    with open(os.path.join(ARTIFACTS_DIR, 'svd_model.pkl'), 'wb') as f:
        pickle.dump(svd_prod, f)
    print(f"Saved: svd_model.pkl")

    np.save(os.path.join(ARTIFACTS_DIR, 'user_factors.npy'), user_factors_prod)
    print(f"Saved: user_factors.npy (Shape: {user_factors_prod.shape})")

    np.save(os.path.join(ARTIFACTS_DIR, 'item_factors.npy'), item_factors_prod)
    print(f"Saved: item_factors.npy (Shape: {item_factors_prod.shape})")
    
    print("\n✅ All artifacts and metrics saved to 'cf-artifacts' folder.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from pymongo import MongoClient
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import time
import os
import json
import re

# --- Configuration ---
MONGO_URI = 'mongodb://localhost:27017/'
DB_NAME = 'RecommendationSystem'
ARTIFACTS_DIR = 'cbf-artifacts' # Folder to store all outputs
# ---------------------

def load_data(client, db_name):
    """Loads all mobile data from MongoDB."""
    db = client[db_name]
    
    mobiles_cursor = db.mobiles.find({})
    mobiles_df = pd.DataFrame(list(mobiles_cursor))
    
    print(f"Loaded {len(mobiles_df)} mobile documents from MongoDB.")
    return mobiles_df

def preprocess_and_create_soup(df):
    """
    Performs feature engineering to combine all mobile specs
    into a single string ("soup") for vectorization.
    """
    print("Engineering features into 'soup'...")
    
    # Fill missing values to avoid errors
    df_clean = df.fillna('')

    # --- Feature Engineering ---
    
    # 1. Bin numerical features into categories
    def bin_battery(cap):
        if not isinstance(cap, (int, float)) or cap == 0:
            return ''
        if cap < 2000: return 'battery_low'
        if cap <= 4500: return 'battery_medium'
        if cap > 4500: return 'battery_high'
        return ''

    def bin_screen(size):
        if not isinstance(size, (int, float)) or size == 0:
            return ''
        if size < 5: return 'screen_small'
        if size <= 6: return 'screen_medium'
        if size > 6: return 'screen_large'
        return ''

    df_clean['battery_bin'] = df_clean['Battery capacity (mAh)'].apply(bin_battery)
    df_clean['screen_bin'] = df_clean['Screen size (inches)'].apply(bin_screen)
    
    # 2. Clean text features
    def clean_text(text):
        text = str(text).lower()
        # Remove "megapixel", "gb", "ghz" and keep the number
        text = re.sub(r'(\d+)\s*(-megapixel|megapixel|mp)', r' \1 ', text)
        text = re.sub(r'(\d+\.?\d*)\s*(-ghz|ghz)', r' \1 ', text)
        # Remove any other non-alphanumeric characters
        text = re.sub(r'[^a-z0-9 ]', ' ', text)
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    # 3. Create the "soup"
    # We combine all important features into one string
    # We use Brand 3 times to give it more weight
    df_clean['soup'] = (
        df_clean['Brand'] + ' ' + df_clean['Brand'] + ' ' + df_clean['Brand'] + ' ' +
        df_clean['Model'] + ' ' +
        clean_text(df_clean['Operating system']) + ' ' +
        clean_text(df_clean['Processor']) + ' ' +
        clean_text(df_clean['Rear camera']) + ' ' +
        df_clean['battery_bin'] + ' ' +
        df_clean['screen_bin']
    )
    
    print("Feature 'soup' created successfully.")
    return df_clean['soup']

def build_artifacts(soup, df):
    """
    Vectorizes the soup, builds the similarity matrix,
    and creates the mappers.
    """
    
    # 1. Vectorize the soup using TF-IDF
    print("Vectorizing text with TF-IDF...")
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(soup)
    
    print(f"Created TF-IDF matrix with shape: {tfidf_matrix.shape}") # (n_mobiles, n_features)

    # 2. Build the Cosine Similarity Matrix
    print("Building Cosine Similarity matrix...")
    # This creates a (n_mobiles, n_mobiles) matrix
    cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
    print(f"Created similarity matrix with shape: {cosine_sim_matrix.shape}")

    # 3. Create mappers
    # We need to map the matrix index (e.g., 0) to the mobile's unique ID (url)
    # and vice-versa.
    item_id_list = list(df['url'])
    item_id_to_index = {url: i for i, url in enumerate(item_id_list)}
    index_to_item_id = {i: url for i, url in enumerate(item_id_list)}
    
    mappers = {
        "item_id_to_index": item_id_to_index,
        "index_to_item_id": index_to_item_id
    }
    
    return cosine_sim_matrix, tfidf_vectorizer, mappers, tfidf_matrix.shape[1]


def save_report(total_items, total_features):
    """Creates and saves a JSON report explaining the model."""
    
    report = {
        "model_type": "Content-Based Filtering (TF-IDF & Cosine Similarity)",
        "dataset_info": {
            "total_items": total_items,
            "total_features_extracted": total_features
        },
        "model_explanation": {
            "what_it_is": "A model that recommends items based on their properties (content). It does not use any user interaction data.",
            "how_it_works": "1. All mobile specs (Brand, OS, Battery, etc.) were combined into a 'soup' of keywords. \n2. TF-IDF was used to find the most important keywords for each phone. \n3. A similarity matrix was built to store a score (0 to 1) of how similar each phone is to every other phone.",
            "what_it_is_used_for": "This model is perfect for 'cold-start' users or for a 'Find Similar Items' feature.",
            "example": "A user is viewing a 'Realme 7 Pro' (which has 'Android 10' and a 'battery_high'). This model can instantly find the top 10 other phones that *also* have 'Android 10' and a 'battery_high', even if they are from a different brand."
        },
        "artifacts_generated": [
            "cbf_similarity_matrix.npy",
            "cbf_mappers.pkl",
            "tfidf_vectorizer.pkl",
            "cbf_report.json"
        ]
    }
    
    report_path = os.path.join(ARTIFACTS_DIR, 'cbf_report.json')
    with open(report_path, 'w') as f:
        json.dump(report, f, indent=4)
    print(f"\nMetrics report saved to {report_path}")

def main():
    # --- 0. Setup ---
    os.makedirs(ARTIFACTS_DIR, exist_ok=True)
    start_time = time.time()
    
    try:
        client = MongoClient(MONGO_URI)
        client.server_info()
    except Exception as e:
        print(f"❌ Error: Could not connect to MongoDB. Is it running? \n   Details: {e}")
        return

    # --- 1. Load Data ---
    mobiles_df = load_data(client, DB_NAME)
    
    # --- 2. Preprocess & Vectorize ---
    soup = preprocess_and_create_soup(mobiles_df)
    
    # --- 3. Build Artifacts ---
    cosine_sim, vectorizer, mappers, n_features = build_artifacts(soup, mobiles_df)

    # --- 4. Save Artifacts ---
    print("\n--- [Saving Artifacts] ---")
    
    # Save the similarity matrix
    np.save(os.path.join(ARTIFACTS_DIR, 'cbf_similarity_matrix.npy'), cosine_sim)
    print("Saved: cbf_similarity_matrix.npy")
    
    # Save the mappers
    with open(os.path.join(ARTIFACTS_DIR, 'cbf_mappers.pkl'), 'wb') as f:
        pickle.dump(mappers, f)
    print("Saved: cbf_mappers.pkl")
    
    # Save the vectorizer (for future use, e.g., vectorizing a search query)
    with open(os.path.join(ARTIFACTS_DIR, 'tfidf_vectorizer.pkl'), 'wb') as f:
        pickle.dump(vectorizer, f)
    print("Saved: tfidf_vectorizer.pkl")

    # --- 5. Save Report ---
    save_report(len(mobiles_df), n_features)
    
    print(f"\n✅ All CBF artifacts built and saved in {time.time() - start_time:.2f} seconds.")

if __name__ == "__main__":
    main()
```
